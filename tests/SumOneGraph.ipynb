```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot the data from busca_largura_tempos.txt, which is separated as "[number]; [time]"
# do the same for busca_profundidade10_tempos.txt, busca_profundidade100_tempos.txt and busca_profundidade_iterativa_tempos.txt

# read the data
largura = pd.read_csv('busca_largura_tempos.txt', sep='; ', engine='python', header=None)
profundidade10 = pd.read_csv('busca_profundidade10_tempos.txt', sep='; ', engine='python', header=None)
profundidade100 = pd.read_csv('busca_profundidade100_tempos.txt', sep='; ', engine='python', header=None)
iterativa = pd.read_csv('busca_profundidade_iterativa_tempos.txt', sep='; ', engine='python', header=None)
```

```python
largura.columns = ['number', 'time']
profundidade10.columns = ['number', 'time']
profundidade100.columns = ['number', 'time']
iterativa.columns = ['number', 'time']
```

```python
def convert_time_to_float(time_str):
    parts = time_str.split(':')
    hours = int(parts[0])
    minutes = int(parts[1])
    seconds_and_milliseconds = parts[2].split('.')
    seconds = int(seconds_and_milliseconds[0])
    total_seconds = hours * 3600 + minutes * 60 + seconds
    return total_seconds
```

```python
for i in range(len(largura)):
    largura['time'][i] = convert_time_to_float(largura['time'][i])
```

```python
# plot the data using plt
plt.figure(figsize=(10, 5))
plt.plot(largura['number'], largura['time'], label='largura')
plt.plot(profundidade10['number'], profundidade10['time'], label='profundidade10')
plt.plot(profundidade100['number'], profundidade100['time'], label='profundidade100')
plt.plot(iterativa['number'], iterativa['time'], label='iterativa')
plt.legend()
plt.xlabel('Goals')
plt.xticks(np.arange(1, 51, 1))
plt.yticks(np.arange(0, 901, 100))
plt.ylabel('Time (s)')
plt.title('Time x Goals for each search method')
plt.show()
```
